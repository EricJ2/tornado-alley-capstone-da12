```python
import pandas as pd

dds_df = pd.read_csv(r"C:\da12\side gig capstone\DisasterDeclarationsSummaries.csv")
dds_df.head()
```

```python
dds_df.columns
```

```python
tornado_df = dds_df[dds_df['incidentType'] == 'Tornado']

tornado_df.head()
```

```python
state_tornado_counts = tornado_df.groupby('state').size().reset_index(name='declaration_count')

state_tornado_counts_sorted = state_tornado_counts.sort_values(by='declaration_count', ascending=False)

state_tornado_counts_sorted.head()
```

#### Tornado declarations over time (1950s-2021)

```python
tornado_df['declarationDate'] = pd.to_datetime(tornado_df['declarationDate'])

tornado_df['year'] = tornado_df['declarationDate'].dt.year
tornado_trends = tornado_df.groupby('year').size()

tornado_trends.plot(kind='line', title='Tornado Declarations Over Time')
```

#### States with the most tornado-related disaster declarations; only 21% found in tornado alley

```python
state_tornado_counts = tornado_df.groupby('state').size().reset_index(name='declaration_count')

state_tornado_counts_sorted = state_tornado_counts.sort_values(by='declaration_count', ascending=False)

state_tornado_counts_sorted.head()
```

```python
import matplotlib.pyplot as plt

top_states = state_tornado_counts_sorted.head(10)

plt.figure(figsize=(10,6))
plt.bar(top_states['state'], top_states['declaration_count'], color='blue')
plt.title('Top 10 States with Most Tornado Declarations')
plt.xlabel('State')
plt.ylabel('Number of Tornado Declarations')
plt.xticks(rotation=45)
plt.show()
```
